```python
# Function to calculate unique species pairs and counts by site
def calculate_species_counts_by_site(df):
    # Group by site
    grouped = df.groupby('site_id')
    site_results = {}

    for site_id, group in grouped:
        # Create a set of unique species pairs for the site
        unique_pairs = group.apply(
            lambda row: tuple(sorted([row['species_a'], row['species_b']])), axis=1
        ).drop_duplicates()

        # Count occurrences of each species in unique pairs
        species_counts = pd.Series(
            [species for pair in unique_pairs for species in pair]
        ).value_counts()

        # Create a new DataFrame for the site's results
        species_counts_df = species_counts.reset_index()
        species_counts_df.columns = ['species', 'observed']

        # Store the results for this site
        site_results[site_id] = species_counts_df

    return site_results

# Example usage
site_species_counts = calculate_species_counts_by_site(significant_df)

# Display the results for each site
for site_id, counts_df in site_species_counts.items():
    print(f"Site ID: {site_id}")
    print(counts_df)
    print("\n" + "-"*40 + "\n")
```

```python
def calculate_species_counts_by_site_as_dataframe(df):
    # Create an empty list to store results
    all_sites_data = []

    # Group by site
    grouped = df.groupby('site_id')

    for site_id, group in grouped:
        # Create a set of unique species pairs for the site
        unique_pairs = group.apply(
            lambda row: tuple(sorted([row['species_a'], row['species_b']])), axis=1
        ).drop_duplicates()

        # Count occurrences of each species in unique pairs
        species_counts = pd.Series(
            [species for pair in unique_pairs for species in pair]
        ).value_counts()

        # Create a DataFrame for this site's results
        species_counts_df = species_counts.reset_index()
        species_counts_df.columns = ['species', 'observed']
        species_counts_df['site_id'] = site_id  # Add site_id as a column

        # Append to the list
        all_sites_data.append(species_counts_df)

    # Combine all site results into a single DataFrame
    return pd.concat(all_sites_data, ignore_index=True)

# Example usage
site_species_counts_df = calculate_species_counts_by_site_as_dataframe(significant_df)

# Display the combined DataFrame
print(site_species_counts_df)
```

```python
# Step 1: Create the `raw_expected` counts for each site
def calculate_raw_expected_per_site(df):
    # Create an empty list to store results
    all_sites_data = []

    # Group by site
    grouped = df.groupby('site_id')

    for site_id, group in grouped:
        # Create unique species pairs (ignoring phenophases) for the site
        group['species_pair'] = group.apply(
            lambda row: tuple(sorted([row['species_a'], row['species_b']])), axis=1
        )
        unique_pairs = group.drop_duplicates(subset=['species_pair'])

        # Count occurrences of each species in unique pairs
        species_counts = pd.concat([
            unique_pairs['species_a'],
            unique_pairs['species_b']
        ]).value_counts().reset_index()

        species_counts.columns = ['species', 'raw_expected']
        species_counts['site_id'] = site_id  # Add site_id as a column

        # Append the results to the list
        all_sites_data.append(species_counts)

    # Combine all site results into a single DataFrame
    return pd.concat(all_sites_data, ignore_index=True)

# Step 2: Calculate `raw_expected` for each site
raw_expected_per_site = calculate_raw_expected_per_site(filtered_pairs)

# Step 3: Merge the `raw_expected` column with the observed results
merged_results = site_species_counts_df.merge(
    raw_expected_per_site, on=['site_id', 'species'], how='left'
)

# Display the final DataFrame
pd.set_option('display.max_rows', None)  # Show all rows
print(merged_results)
```

```python
# Step 1: Group by site and calculate the required values
def calculate_expected_per_site(df):
    # Group by site_id
    results = []

    for site_id, group in df.groupby('site_id'):
        # Step 1: Calculate total raw expected counts for the site
        total_raw_expected = group['raw_expected'].sum()

        # Step 2: Calculate the proportion of unique pairings for each species in the site
        group['proportion'] = group['raw_expected'] / total_raw_expected

        # Step 3: Scale the proportions to match the total observed significant pairings
        total_observed = group['observed'].sum()
        group['expected'] = group['proportion'] * total_observed

        # Append the updated group to results
        results.append(group)

    # Combine the updated groups back into a single DataFrame
    return pd.concat(results)

# Apply the function to calculate per-site expected values
merged_results_with_expected = calculate_expected_per_site(merged_results)

# Display the updated DataFrame
pd.set_option('display.max_rows', None)  # Show all rows
print(merged_results_with_expected)
```

```python
from scipy.stats import chisquare

# Step 1: Function to calculate chi-square per site
def calculate_chi_square_per_site(df):
    results = []

    for site_id, group in df.groupby('site_id'):
        # Step 1.1: Calculate chi-square contribution for each species in the site
        group['chi_square_contribution'] = (
            (group['observed'] - group['expected']) ** 2 / group['expected']
        )

        # Step 1.2: Calculate overall chi-square statistic and p-value for the site
        chi_square_stat = group['chi_square_contribution'].sum()
        degrees_of_freedom = len(group) - 1  # Number of species - 1
        chi_square_results = chisquare(
            f_obs=group['observed'], 
            f_exp=group['expected']
        )
        p_value = chi_square_results.pvalue

        # Step 1.3: Add site-level results to the group
        group['chi_square_stat'] = chi_square_stat
        group['degrees_of_freedom'] = degrees_of_freedom
        group['p_value'] = p_value

        # Append the updated group to results
        results.append(group)

    # Combine all site-level DataFrames
    return pd.concat(results)

# Step 2: Apply the function to calculate per-site chi-square statistics
merged_results_with_chi_square = calculate_chi_square_per_site(merged_results_with_expected)

# Step 3: Display the updated DataFrame with per-site chi-square results
pd.set_option('display.max_rows', None)  # Show all rows
merged_results_with_chi_square
```
